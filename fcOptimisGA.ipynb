**Problem statement:** Implement a GA for function optimisation: identify the optimal point of f(x), where:
- f : \[a, b\]^n -> R, f(x) = f(x1, x2, x3, ..., xn)
Possible functions:
- f(x) = sum(xi), i = 1,2,...,n, \[a, b\] = \[-5, 5\]
- f(x) = sum(xi^2 + xi * cos(2 * Pi * xi)), i = 1, 2, ..., n, \[a, b\] = \[-5, 5\]
- f(x) = 20 + sum(xi^2 - 10 * cos(2 * Pi * xi), i = 1, 2, ..., n, \[a, b\] = \[-5, 5\]
- f(x) = sum(xi^4 - 16 * (xi^2) + 5 * xi) / 2, i = 1, 2, ..., n, \[a, b\] = \[-5, 5\]




**Function ladscape**

Define the function and plot it in order to observe its shape/aspect and optimal points

```python
# define the function
import math

MIN = -5
MAX = 5

def fcEval(x):    
    val = sum([(xi * xi + xi) * math.cos(2 * 3.14 * xi) for xi in x])
    # val = sum(xi ** 2 for xi in x)
    # val = 20 + sum(xi ** 2 - 10 * np.cos(2 * np.pi * xi) for xi in x)
    # val = (0 - 20) * math.exp(-0.2 * math.sqrt(0.5 * x[0] * x[0])) - math.exp(0.5 * math.cos(2 * 3.14 * x[0])) + math.e + 20
    # val = sum(xi ** 4 - 16 * (xi ** 2) + 5 * xi for xi in x) / 2
    return val


# define a generator of values
from random import uniform

def generateNewGene(lim1, lim2):
    return uniform(lim1, lim2)

#  plot the 1D function (n = 1) (see how the search space looks)
import matplotlib.pyplot as plt

noDim = 1
xref =  [[generateNewGene(MIN, MAX)  for _ in range(noDim)] for _ in range(0, 1000)]
xref.sort()
yref = [fcEval(xi) for xi in xref]   

plt.ion()
plt.plot(xref, yref, 'b-')
plt.xlabel('x values')
plt.ylabel('y = f(x) values')
plt.show()

```
